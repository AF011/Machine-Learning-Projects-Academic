```python
'''
Display the structure of the dataset and identify the numeric
Columns and calculate mean , median, mode, min, max, Range
And IQR with q1 and q3 and display all the values for each
columns and Draw the boxplot and histogram for each
column and interpret the data.
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv('1st_sem_marks.csv')
Dataset
```

```python
#Dimensions of the Data
Dataset.shape
```

```python
#Structure of the Data
Dataset.info()
```

```python
#Average Marks in Each Subject
Marks=Dataset.iloc[:, 1:]
Marks.mean()
```

```python
#Median Value in each Subject
Marks.median()
```

```python
#Mode Value of each Subject
Marks.mode()
```

```python
#Quartiles
#Q1
q1=Marks.quantile(0.25)
q1
```

```python
#q3
q3=Marks.quantile(0.75)
q3
```

```python
#Quartile Difference between q3 and q1.
q=q3-q1
q
```

```python
#Histogram
Marks.hist()
```

```python
#Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(Marks)
plt.xticks(rotation=90)
plt.tight_layout()
plt.xlabel('Subjects')
plt.ylabel('')
plt.title('Marks Variation of Students in each Subject')
plt.show()
```
