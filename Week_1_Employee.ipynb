```python
#Employee Randomly Generated Dataset
import pandas as pd
import numpy as np
emp_details={
    "Emp_ID":np.arange(1, 101),
    "Age":np.random.randint(20, 65, size=100),
    "Salary":np.random.randint(14000, 120000, size=100),
    "Experience":np.random.randint(1, 20, size=100),
    "Performance_Score":np.random.randint(1, 10, size=100)
}
emp_details=pd.DataFrame(emp_details)
emp_details.to_csv('Employee_Details_Random.csv', index=False)
emp_details.head(n=10)
```
